```python
#exploring Label encoder using sklearn
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import pandas as pd
DataF = ('Mumbai', 'Delhi', 'Hyderabad', 'Chennai', 'Calcutta', 'Delhi', 'Hyderabad')
Cit = pd.DataFrame(DataF, columns=['Cities'])
Cit
```

```python
lbEnd = LabelEncoder()
Cit['Cities_Encoded'] = lbEnd.fit_transform(Cit['Cities'])
Cit
```

```python
#exploring One Hot Encoder
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown='ignore')
e_df = pd.DataFrame(enc.fit_transform(Cit[['Cities']]).toarray())

Cit = Cit.join(e_df)
Cit
```

```python
#Classification of Assignment Question

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#printing the dataset
db = pd.read_excel('Acoustic_Extinguisher_Fire_Dataset.xlsx', sheet_name='A_E_Fire_Dataset')
db.head()
```

```python
Fl = pd.get_dummies(db.FUEL)
Fl
```

```python
db = pd.concat([db, Fl], axis="columns")
db
```

```python
db.drop('FUEL', inplace=True, axis=1)
db
```

```python
sts = db.pop('STATUS')
```

```python
db.insert(9, 'STATUS' ,sts)
db
```

```python
#extracting X and Y values
X = db.iloc[:, 0:9]
Y = db.iloc[:, 9]
```

```python
#splitting of training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.3, random_state=4)
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='rbf')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
print(*y_pred[0:10])
#0 shows that fire was not extinguished whereas 1 shows that it was extinguised
```

```python
from sklearn import metrics
acc = metrics.accuracy_score(y_test, y_pred)
print("Accuracy of our model is: ",acc*100)
```

```python
print("Precision:", metrics.precision_score(y_test, y_pred))
```

```python
print("Recall:", metrics.recall_score(y_test, y_pred))
```
